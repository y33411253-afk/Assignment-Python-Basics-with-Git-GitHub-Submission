**Q1 Arithmetic and Assignment Operations**

```python
a=10
b = 44

print("Addition: ", a + b)
print("Subtraction: ", a - b)
print("Multiplication: ", a * b)
print("Division: ", a/b )
print("Floor Division: ", a//b )
print("Modulus: ", a%b )
print("Exponentiation: ", a**b )

#assignment operators
a += 2
print("After += : ", a )
a -= 3
print("After -= : ", a)
a *= 4
print("After *= : ", a )
a/= 5
print("After /= : ", )
```

**Q2 Comparison and Logical operators**

```python
x = int(input("Enter first num: "))
y = int(input("Enter second num: "))

#comparison operators
print("x == y: ",x ==y )
print("x!=y: ", x !=y )
print("x > y: ", x > y)
print("x < y: ", x < y )
print("x >= y: ", x >= y )
print("x <= y: ", x <= y )

#logical operators
print("x > 0 and y >0: ", x > 0 and y > 0)
print("x > 0 or y > 0: ", x > 0 or y > 0)
print("not(x>y): ", not(x > y))
```

**Q3 Bitwise Operators**

```python
p = 1
q = 2
print("p & q (and): ", p & q  )
print("p | q (or ): ", p | q )
print("p ^ q (xor): ", p ^ q )
print("~p (not): ", ~p )
print("p << 1 (left shift): ", p<< 1 )
print("p >> 1 (right shift): ", p >> 1 )
```

**Q4 Identity and Membership Operators**

```python
nums = [1, 2, 3, 4, 5]
a = nums
b = [1, 2, 3, 4, 5]

# Identity operators
print("a is numbers: ", a is nums)
print("b is numbers:", b is nums)
print("a is not b: ", a is not b)

# Membership operators
print("3 in numbers:", 3 in nums)
print("10 not in numbers:", 10 not in nums)
```

**Q5 String operators**

```python
sentence = "Hello World!!!"

#length
print("Length:", len(sentence))

#uppercase and lowercase
print("Uppercase: ", sentence.upper() )
print("Lowercase: ", sentence.lower() )

#finding word
print("'World' in sentence:", "World" in sentence )

#replacing
print("Replace 'World' with 'Universe': ", sentence.replace("World", "Universe" ))

#slicing
print("Slice (first 7 chars): ", sentence[: 7] )

#reverse
print("Reversed: ", sentence[::-1])
```
